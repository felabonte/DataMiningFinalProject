```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import Grouper
from pandas import DataFrame
```

```python
hdout = pd.read_csv('HistDebtOutstanding.csv') 
hdout
```

```python
list(hdout.columns)
```

```python
# Clean Data
hdout_clean = hdout.dropna()
hdout_clean
```

```python
hdout_clean.dtypes
```

```python
hdout_clean['Record Date']= pd.to_datetime(hdout_clean['Record Date'])
hdout_clean.info()
```

```python
hdout_clean = hdout_clean.set_index('Record Date')
hdout_clean.head(3)
```

```python
hdout_clean.index
```

```python
# Add columns with year, month, and weekday name
hdout_clean['Year'] = hdout_clean.index.year
hdout_clean['Month'] = hdout_clean.index.month
hdout_clean['Weekday Name'] = hdout_clean.index.day
# Display a randdom sampling of 5 rows
hdout_clean.sample(5, random_state=0)
```

```python
hdout_clean.plot.line(x='Year', y='Debt Outstanding Amount')
```
