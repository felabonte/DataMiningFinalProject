```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import Grouper
from pandas import DataFrame
```

```python
fdrates = pd.read_csv('AvgInterestRate.csv') 
fdrates
```

```python
list(fdrates.columns)
```

```python
# Clean Data
rslt_df = fdrates.dropna()
rslt_df
```

```python
rslt_df.dtypes
```

```python
rslt_df['Record Date']= pd.to_datetime(rslt_df['Record Date'])
rslt_df.info()
```

```python
rslt_df = rslt_df.set_index('Record Date')
rslt_df.head(3)
```

```python
rslt_df.index
```

```python
# Add columns with year, month, and weekday name
rslt_df['Year'] = rslt_df.index.year
rslt_df['Month'] = rslt_df.index.month
rslt_df['Weekday Name'] = rslt_df.index.day
# Display a randdom sampling of 5 rows
rslt_df.sample(5, random_state=0)
```

```python
# list of values of 'Security Description' column
security_types = rslt_df['Security Description'].unique()
security_type_set = security_types.tolist()
security_type_set
```

```python
# list of values of 'Security Description' column
security_descr = rslt_df['Security Type Description'].unique()
security_descr_set = security_types.tolist()
security_descr_set
```

```python
sns.set(rc={'figure.figsize':(11, 4)})
for sec_type in security_type_set:
    df = rslt_df[rslt_df['Security Description'] == sec_type] 
    df['Average Interest Rate Amount'].plot(linewidth=0.5)
    plt.xlabel('Date')  
    plt.ylabel('Rate')  
  
    # displaying the title 
    plt.title("Time Series " + sec_type) 
    plt.show()
    
```

```python
sns.set(rc={'figure.figsize':(11, 4)})
for sec_type in security_descr_set:
    df = rslt_df[rslt_df['Security Type Description'] == sec_type] 
    df['Average Interest Rate Amount'].plot(linewidth=0.5)
    plt.xlabel('Date')  
    plt.ylabel('Rate')  
  
    # displaying the title 
    plt.title("Time Series " + sec_type) 
    plt.show()
    
```
