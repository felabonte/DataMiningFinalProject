```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering, KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Importlibraries

# Step 2: Load the CSV data into DF
df = pd.read_csv('MSPD_SumSecty_20010131_20231031.csv')

# Step 3: Preprocess the data

# Encode categorical variables
label_encoder = LabelEncoder()
df['Security Type Description'] = label_encoder.fit_transform(df['Security Type Description'])
df['Security Class Description'] = label_encoder.fit_transform(df['Security Class Description'])

# Standardize numerical features to Peform Agg Cluster and K-Means by 3 atributes.
scaler = StandardScaler()
numerical_cols = ['Debt Held by the Public (in Millions)', 'Intragovernmental Holdings (in Millions)', 'Total Public Debt Outstanding (in Millions)']
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# Step 4: Perform Agglomerative Clustering
agg_clustering = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
df['Agg_Cluster'] = agg_clustering.fit_predict(df[numerical_cols])

# Step 5: Perform K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=0)
df['KMeans_Cluster'] = kmeans.fit_predict(df[numerical_cols])

# Step 6: Visualize the results

# Use PCA to reduce dimensionality for visualization
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df[numerical_cols])

# Plot Agglomerative Clustering
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=df['Agg_Cluster'], cmap='viridis')
plt.title('Agglomerative Clustering')

# Plot K-Means Clustering
plt.subplot(1, 2, 2)
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=df['KMeans_Cluster'], cmap='viridis')
plt.title('K-Means Clustering')

plt.show()
```
